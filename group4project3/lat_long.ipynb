```python
#Dependencies
import pandas as pd
import numpy as np
import requests
import json
```

```python
from config import gkey
ta_pd = pd.read_csv("tourist_attraction.csv")
ta_pd.head()
```

```python
params = {"key": gkey}
#Loop through the cities and run a lat long search
for index, row in ta_pd.iterrows():
    base_url ="https://maps.googleapis.com/maps/api/geocode/json"
    
    address = row ['address']
    #update address key value
    params ['address'] = f"{address}"
    lat_lng = requests.get(base_url, params=params)
# print(lat_lng.url)
    lat_lng = lat_lng.json()
    ta_pd.loc[index, "lat"] = lat_lng["results"][0]["geometry"]["location"]["lat"]
    ta_pd.loc[index, "lon"] = lat_lng["results"][0]["geometry"]["location"]["lng"]
ta_pd.head()
```

```python
ta_pd.to_html('top_attractions')
```
